```python
import numpy as np
import pandas as pd
```

```python
num_pts = 50
x1 = 2*np.random.rand(num_pts) - 1
x1.shape = (len(x1),1)
x2 = 2*np.random.rand(num_pts) - 1
x2.shape = (len(x2),1)
y = 1 - x1 - x2 + 0.1*np.random.randn(num_pts,1)
y.shape = (len(y),1) 
```

```python
data = np.concatenate((x1,x2,y),axis = 1)
```

```python
np.savetxt('test.csv',data,delimiter=',')
```
